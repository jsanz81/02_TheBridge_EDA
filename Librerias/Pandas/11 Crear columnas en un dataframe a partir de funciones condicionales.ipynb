## Librería Pandas
Como crear nuevas clumnas a partir de funciones condicionales

```python
import pandas as pd
```

```python
import numpy as np
```

```python
array_gaus = np.random.normal(3,1,1000000)
df = pd.DataFrame(data=array_gaus, columns=["Valores"])
```

```python
df
```

```python
def supinf(x):
    if x>3:
        z = "Valor mayor que la media"
    else:
        z = "Valor menor que la media"
    return z
```

```python
df["Chequeo"]=df["Valores"].apply(supinf)
```

```python
df
```

```python
df["Chequeo2"] = np.where(df["Valores"]>3,"Valor mayor que la media","Valor menor que la media")
```
