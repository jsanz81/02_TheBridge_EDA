* En que fechas fue lanzada cada pelicula de ``Salma Hayek`` desde  1990 al presente en Mexico

```python
peliculas_mexicanas=estreno[["title", "date"]][(estreno.country=="Mexico")&(estreno.year>=1990)]
peliculas_mexicanas

salma=cast[cast.name=="Salma Hayek"]

salma.reset_index(drop=True)
#  merge peliculas_mexicanas y salma
```

```python
salma_mexico=pd.merge(peliculas_mexicanas,salma, left_on="title", right_on="title", how="inner")[["title","date"]].set_index("title")
```

```python
salma_mexico.sort_values("date")
```

```python
from datetime import datetime

peliculas_USA=estreno[estreno.country=="USA"]
tom_cruise=cast[cast.name=="Tom Cruise"]
```

```python
#merge peliculas USA con estrenos de tom cruise por título
tom_USA=pd.merge(peliculas_USA, tom_cruise, left_on="title", right_on="title", how="inner")[["title", "date"]].set_index("title")

tom_USA["fecha"]=pd.to_datetime(tom_USA["date"])
#crear columna "dia", a partir de la columna "date"
```

```python
tom_USA.drop(columns=["date"], inplace=True)
```
